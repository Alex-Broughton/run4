```python
def getData(repo, col, fluxkey="base_CircularApertureFlux_70_0_instFlux", shapekey="base_SdssShape_xx"):
    #fluxkey="base_CircularApertureFlux_70_0_instFlux", shapekey="base_SdssShape_xx"):
    # Pulls out data from run and returns BF table
    
    subbutler = Butler(repo,collections=col)
    subregistry = subbutler.registry
    
    datasetRefs = list(subregistry.queryDatasets(
        datasetType="gridSpotSrc",
        collections=col))

    tab = []
    for i, aref in enumerate(datasetRefs):

        # postISRCCD
        md = subbutler.get("raw",dataId=aref.dataId).getMetadata()
        exptime = md["EXPTIME"]

        src = subbutler.get("gridSpotSrc",dataId=aref.dataId)
        mdsrc = src.getMetadata()
        x0 = mdsrc['GRID_X0']
        y0 = mdsrc['GRID_Y0']
        
        minSizeX = 4.0
        maxSizeX = 7.0
        minSizeY = 4.0
        maxSizeY = 7.0
        minFluxRatio = 0.80
        maxFlux = np.nanmax(src['base_SdssShape_instFlux'])
        select = src['base_SdssShape_instFlux'] > maxFlux * minFluxRatio
        spotCatalog  = src.subset(select)
        select = ((src['base_SdssShape_xx'] >= minSizeX) & (src['base_SdssShape_xx'] <= maxSizeX) & 
                (src['base_SdssShape_yy'] >= minSizeY) & (src['base_SdssShape_yy'] <= maxSizeY))


        src  = src.subset(select)
        

        maxradius = 10.*np.mean([mdsrc["GRID_XSTEP"], mdsrc["GRID_YSTEP"]])
        distances = np.sqrt((src["spotgrid_x"] - x0)**2 + (src["spotgrid_y"] - y0)**2)
        mask1 = (distances <= maxradius)

        Iperr = src["base_PsfFlux_instFluxErr"]
        Ip = src["base_PsfFlux_instFlux"]    
        Ierr = src["{}Err".format(fluxkey)]
        I = src["{}".format(fluxkey)]            

        mask2 = (src["base_PixelFlags_flag"] == False) & \
                (src["base_PsfFlux_flag_edge"] == False) & \
                (src["base_CircularApertureFlux_70_0_flag_apertureTruncated"] == False) & \
                (src["base_ClassificationExtendedness_flag"] == False) & \
                (src["base_PixelFlags_flag"] == False) & \
                (src["base_PixelFlags_flag_offimage"] == False) & \
                (src["base_PixelFlags_flag_edge"] == False) & \
                (src["base_PixelFlags_flag_interpolated"] == False) & \
                (src["base_PixelFlags_flag_saturated"] == False) & \
                (src["base_PixelFlags_flag_cr"] == False) & \
                (src["base_PixelFlags_flag_bad"] == False) & \
                (src["base_PixelFlags_flag_suspect"] == False) & \
                (src["base_PixelFlags_flag_interpolatedCenter"] == False) & \
                (src["base_PixelFlags_flag_saturatedCenter"] == False) & \
                (src["base_PixelFlags_flag_crCenter"] == False) & \
                (src["base_PixelFlags_flag_suspectCenter"] == False) & \
                (Iperr/Ip<0.01 ) & \
                (Ierr/I<0.01)


        mask3 = (src['spotgrid_index'] >= 0) #& (src["base_SdssShape_xx"] <= 7)
        
        src = src[mask1 & mask3]
        
        maxADCval = []
        s = 15
        if len(src) > 0:
            postisr = subbutler.get("postISRCCD",dataId=aref.dataId)
            image = postisr.getImage().getArray()
            for i,pt in enumerate(src["spotgrid_index"]):
                x = int(src['spotgrid_y'][i]) # The coordinate systems are flipped
                y = int(src['spotgrid_x'][i])
                xmin = max(0,x-s)
                xmax = min(image.shape[1]-1, x+s)
                ymin = max(0,y-s)
                ymax = min(image.shape[0]-1, y+s)
                maxADCval.append(np.max(image[xmin:xmax,ymin:ymax]))
        
            
        if len(src) > 0:
            tab.append(
                {
                    "index": i,
                    "dataId": aref.dataId,
                    "exptime": exptime,
                    "numspots" : len(src),
                    "spot_indices": src['spotgrid_index'],
                    "spotgrid_x": src['spotgrid_x'],
                    "spotgrid_y": src['spotgrid_y'],
                    "shape": src[shapekey],
                    "shapemean": np.nanmean(src[shapekey]),
                    "shapemedian": np.nanmedian(src[shapekey]),
                    "shapeerr": np.nanstd(src[shapekey]),
                    "flux": src[fluxkey],
                    "fluxmean": np.nanmean(src[fluxkey]),
                    "fluxmedian": np.nanmedian(src[fluxkey]),
                    "fluxerr": np.nanstd(src[fluxkey]),
                    "maxADCval": maxADCval
                }
            )

    t = Table(tab)
    t_sorted = t[np.argsort(t['exptime'])]

    return t_sorted

```
